```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AAA.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum() #It return the number of missing values in dataset
```

```python
df.columns
```

```python
x=df[['ENGINESIZE']].values
```

```python
y=df[['CO2EMISSIONS']].values
```

```python
plt.scatter(x,y,marker='+',color="red")
plt.xlabel('ENGINESIZE')
plt.ylabel('CO2EMISSIONS')
plt.title("co2")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)#use for test and train.
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
plt.plot(df['ENGINESIZE'],reg.predict(df[['ENGINESIZE']]))# plot function use for data visualization
```

```python
reg.score(x_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_#shows the point where the estimated regression line crosses the 𝑦 axis
```
